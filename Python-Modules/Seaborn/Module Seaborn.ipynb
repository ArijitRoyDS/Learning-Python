# Module Seaborn

Seaborn is an amazing visualization library for statistical graphics plotting in Python. It provides beautiful default styles and color palettes to make statistical plots more attractive. It is built on the top of matplotlib library and also closely integrated to the data structures from pandas.
Seaborn aims to make visualization the central part of exploring and understanding data. It provides dataset-oriented APIs, so that we can switch between different visual representations for same variables for better understanding of dataset.

Installation: pip install seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 2-D Scatter Plots

```python
a = sns.load_dataset("flights")
a.head(5)
```

```python
# 2-D Scatter Plots

fig, ax = plt.subplots(figsize=(8,4), dpi=100)
sns.scatterplot(x = "passengers", y = "month", hue = "year", data = a, ax=ax)
```

# 2-D Line Plot

```python
b = sns.load_dataset("tips")
b.head(5)
```

```python
# 2-D Line Plot with Legend

fig, ax = plt.subplots(figsize=(8,4), dpi=100)
sns.lineplot(x = "time", y = "tip", hue = "sex", data = b, ax=ax)
```

# 2-D Bar Plot

```python
# 2-D Bar Plot with Legend
fig, ax = plt.subplots(figsize=(8,4), dpi=100)
sns.barplot(x = "day", y = "total_bill", hue = 'sex', data = b, ax=ax)
```

# 2-D Box Plot

```python
# 2-D Box Plot with Legend
fig, ax = plt.subplots(figsize=(12,4), dpi=200)
sns.boxplot(x = "day", y = "total_bill", hue = 'smoker', data = b, ax=ax)
```

# Histogram

```python
c = sns.load_dataset("titanic")
c.head(5)
```

```python
# 2-D Histogram
fig, ax = plt.subplots(figsize=(8,3), dpi=100)
sns.histplot(c[['age']], binwidth = 5, ax=ax)
```

# Kernel Density Estimate (KDE) Plot

```python
# 2-D KDE Plot
fig, ax = plt.subplots(figsize=(8,3), dpi=100)
sns.kdeplot(data=b, x="total_bill", hue="time", ax=ax)
```

# Distribution or Dist Plot

```python
# Dist Plot
fig, ax = plt.subplots(figsize=(8,3), dpi=100)
sns.distplot(c['age'], kde = True, bins = 20, ax=ax)
```

# FacetGrid

```python
p = sns.load_dataset("iris")
q = sns.FacetGrid(p, col = "species")
q.map(plt.hist, "sepal_length")
```

```python
#sns.set(style = "darkgrid")
p = sns.load_dataset("flights")
q = sns.PairGrid(p)
q.map(plt.scatter)
```

```python
sns.set(style = "white", color_codes = True)
a = sns.load_dataset('tips')
sns.boxplot(x = "day", y = "total_bill", data = a)
sns.despine(offset = 10, trim = True)
```
