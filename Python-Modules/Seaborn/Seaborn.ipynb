## Seaborn

Seaborn is an amazing visualization library for statistical graphics plotting in Python. It provides beautiful default styles and color palettes to make statistical plots more attractive. It is built on the top of matplotlib library and also closely integrated to the data structures from pandas.
Seaborn aims to make visualization the central part of exploring and understanding data. It provides dataset-oriented APIs, so that we can switch between different visual representations for same variables for better understanding of dataset.

#### additional Features
 
 - Column Wrap 
 - Reference Line
 - custom function

```python
def annotate(data, **kwargs):
    n = len(data)
    ax = plt.gca()
    ax.text(.1, .6, f'N={n}', transform=ax.transAxes)
    
g = sns.FacetGrid(data=tips, col='size', col_wrap=3, height=4, aspect=0.8)
g.map(sns.scatterplot, 'total_bill', 'tip')
g.refline(y=tips['tip'].median())
g.map_dataframe(annotate)

g.set_axis_labels('Total Bill ($)', 'Tip ($)')
g.set_titles(col_template="Table Size {col_name}")
g.set(xlim=(0,60), ylim=(0,12), xticks=[10,30,50], yticks=[2,6,10])
g.tight_layout()

# g.savefig('facetgrid.png')
```

__Note:__ transAxes indicates that the coordinates are given relative to the axes bounding box, with (0, 0) being the lower left of the axes and (1, 1) the upper right.

### Combining Multiple Views

__Figure-Level__
- `jointplot()` - belongs to `JointGrid()` object
    - `JointGrid()` - Grid for drawing a bivariate plot with marginal univariate plots.

```python
sns.jointplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
```

```python
# sns.JointGrid(height=4, ratio=2, space=0.05)
g = sns.JointGrid(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
g.plot(sns.scatterplot, sns.histplot, alpha=.7, edgecolor=".2", linewidth=.5)
# g.refline(x=200, y=47)
```

```python
# To pass arguments to each plot functions
g = sns.JointGrid(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.histplot, kde=True)
```

```python
# Access ax_joint, ax_marg_x, and ax_marg_y attributes to plot each axes separately
g = sns.JointGrid(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
sns.scatterplot(x=penguins['flipper_length_mm'], y=penguins['bill_length_mm'], ax=g.ax_joint)
sns.histplot(x=penguins['flipper_length_mm'], ax=g.ax_marg_x)
sns.kdeplot(y=penguins['bill_length_mm'], ax=g.ax_marg_y)
```

- `pairplot()` - belongs to `PairGrid()` object
    - `PairGrid()` - Subplot grid for plotting pairwise relationships in a dataset.

```python
sns.pairplot(data=penguins, hue='species', height=3.5, aspect=1.2)
```

```python
g = sns.PairGrid(data=penguins, hue='species')

g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.scatterplot)

g.add_legend()
```

##### Continuity with Line plots

```python
np.random.seed(10)
d1 = np.random.randn(500).cumsum()
g = sns.relplot(data=d1, kind='line', height=10, aspect=2)
```

```python
np.random.seed(20)
d2 = pd.DataFrame(np.random.randn(500,2).cumsum(axis=0), columns=['x','y'])
sns.relplot(data=d2, kind='line', x='x', y='y', height=10, aspect=2, sort=False)
```

## Categorical Data

- relationships between categorical and numerical data

- `catplot()`

##### Categorical scatterplots:

- `stripplot()` (with `kind="strip"` default)
- `swarmplot()` (with `kind="swarm"`)

##### Categorical distribution plots:

- `boxplot()` (with `kind="box"`)
- `violinplot()` (with `kind="violin"`)
- `boxenplot()` (with `kind="boxen"`)

##### Categorical estimate plots:

- `pointplot()` (with `kind="point"`)
- `barplot()` (with `kind="bar"`)
- `countplot()` (with `kind="count"`)

`StripPlot` adjusts the position of points on the categorical axis with some/without **jitters**

```python
sns.catplot(data=tips, 
            x='day', 
            y='total_bill', 
            # jitter=False,
            kind='strip', 
            height=8, aspect=1.2)
```

`SwarmPlot` adjust the points along the categorical axis that prevents overlapping

```python
sns.catplot(data=tips, 
            x='day', 
            y='total_bill',
            hue='smoker',
            kind='swarm',
            height=8, aspect=1.2)
```

```python
sns.catplot(data=tips, 
            x='smoker', y='tip', 
            order=['Yes', 'No'], 
            height=8, aspect=1.2)
```

##### Distributions within Categories

```python
# boxplot
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```
